```python
#ReadMe:  ,       ,        (-1)

import math
import matplotlib.pyplot as plt

class Shapes:

    def __init__(self, n_angles, angles, sides):
        self.n_angles = n_angles    #    ( );
        self.angles = angles        #     ([90, 90, 90, 90]);
        self.sides = sides          #   ([10, 10, 10, 10]) ( 1-   2-, 2-  3-, 3-  4-, 4-  1-);
        self.validate_shape()

    def validate_shape(self):
        if self.n_angles == 0:     # -,    
            return
        if any(angle <= 0 or angle >=360 for angle in self.angles):                 #,     0 <= angle <=360
            raise ValueError("All angles must be more ore equal than 0 and less or equal than 360!")
        if not isinstance(self, Triangle): #  ,    ,       ,    2 
            if len(self.sides) != self.n_angles and len(self.sides):             #,        ,   
                raise ValueError("The number of sides must be equal to n_angles!")
            if not all(side < sum(self.sides) - side for side in self.sides):         #  :       
                raise ValueError('Every side must be less than the sum of other sides!')

    def get_perimetr(self):         #    
        return sum(self.sides)

    def get_info(self):
        return {'Number of angles ' : self.n_angles,
                'Angles ' : self.angles,
                'Length of the sides ' : self.sides,
                'Perimetr ' : self.get_perimetr()
                }
class Circle(Shapes):

    def __init__(self, radius):
        super().__init__(0, [], [radius])
        if radius <= 0:
            raise ValueError("The radius must be more than 0!")
        self.name = "Circle"     #  name
        self.radius = radius

    def get_sq(self):    # 
        return math.pi * self.sides[0] * self.sides[0]

    def get_perimetr(self):    # 
        return 2 * math.pi * self.sides[0]

    def set_angles(self, angles):   # ,      
        raise ValueError("The circle has no angles!")

    def set_sides(self, sides):  # 
        if sides[0] <= 0 or len(sides) != 1:
            raise ValueError("The circle must have one positive radius!")
        self.sides = sides

    def get_info(self):  # -
        return {'Name ': self.name,
                'Radius ': self.sides[0],
                'Perimetr ': self.get_perimetr(),
                'Square ' : self.get_sq()
                }

    def draw(self):
        circle = plt.Circle((0, 0), self.radius, color='blue', fill=True, alpha=0.5)
        fig, ax = plt.subplots()
        ax.add_artist(circle)

        ax.set_xlim(-self.radius - 1, self.radius + 1)
        ax.set_ylim(-self.radius - 1, self.radius + 1)
        ax.set_aspect('equal', adjustable='box')  #   
        plt.title("Circle with radius " + str(self.radius))
        plt.grid()
        plt.show()

class Triangle(Shapes):
    def __init__(self, angles=None, sides=None):
        if sides is None or len(sides) != 3:
            raise ValueError("Triangle must have 3 sides!")
        self.sides = sides

        if angles is not None and len(angles) == 3:
            self.angles = angles
        else:
            self.angles = self.calculate_angles_from_sides(sides)

        super().__init__(3, self.angles, self.sides)
        self.name = "Triangle"

    def calculate_angles_from_sides(self, sides):
        a, b, c = sides
        #      
        angle_A = math.degrees(math.acos((b ** 2 + c ** 2 - a ** 2) / (2 * b * c)))
        angle_B = math.degrees(math.acos((a ** 2 + c ** 2 - b ** 2) / (2 * a * c)))
        angle_C = 180 - angle_A - angle_B
        return [angle_A, angle_B, angle_C]

    def get_sq(self):  #  
        # 3 ,               :1    (  ),   )
        if self.angles[0] == 90:
            return 0.5 * self.sides[1] * self.sides[2]
        if self.angles[1] == 90:
            return 0.5 * self.sides[0] * self.sides[2]
        if self.angles[2] == 90:
            return 0.5 * self.sides[1] * self.sides[0]
        if len(self.sides) == 2 and len(self.angles) == 1: #           :2    (     )
            a = self.sides[0]
            b = self.sides[1]
            alpha = math.radians(self.angles[0])
            return 0.5 * a * b * math.sin(alpha)
        s = self.get_perimetr() / 2    #                               :3    (  ) 
        return math.sqrt(s * (s - self.sides[0]) * (s - self.sides[1]) * (s - self.sides[2]))

    def set_angles(self, angles):  #  
        if len(angles) == 2:
            if all(0 < angle < 180 for angle in angles):   #   2 ,   
                third_angle = 180 - sum(angles)
                if 0 < third_angle < 180:
                    self.angles = angles + [third_angle]
                else:
                    raise ValueError("Calculated third angle is not valid!")
            else:
                raise ValueError("Angles must be greater than 0 and less than 180 degrees!")
        elif len(angles) == 3:
            if all(0 < angle < 180 for angle in angles):
                self.angles = angles
            else:
                raise ValueError("Angles must be greater than 0 and less than 180 degrees!")
        else:
            raise ValueError("The triangle must have 3 angles!")
        self.validate_shape()

    def set_sides(self, sides):  #  
        if len(sides) == 1 or any(side <= 0 for side in sides): #!!!!!!!!!!!!!!!!!
            raise ValueError("The triangle must have 3 positive sides!")
        if len(sides) == 2:
            if self.angles[0] == 60 and self.angles[1] == 60 and self.angles[2] == 60:  #  
                sides.append(self.sides[0])
                self.sides = sides
            else:
                betta = math.radians(self.angles[0])
                sides.append(math.sqrt(sides[0] * sides[0] + sides[1] * sides[1] - 2 * sides[0] * sides[1] * math.cos(betta)))
                self.sides = sides
        #print(self.sides)
        self.angles = self.calculate_angles_from_sides(sides)
        self.sides = sides
        self.validate_shape()
        self.validate_triangle()

    def validate_triangle(self):
        angles_rad = [math.radians(angle) for angle in self.angles]
        ratio = [self.sides[i] / math.sin(angles_rad[i]) for i in range(len(self.sides))]
        if not all(math.isclose(ratio[i], ratio[0], rel_tol=1e-9) for i in range(1, len(ratio))):
            raise ValueError("The angles don't match with the sides")
        if len(self.sides) != self.n_angles and len(self.sides):  # ,        ,   
            raise ValueError("The number of sides must be equal to n_angles!")
        if not all(side < sum(self.sides) - side for side in self.sides):  #   :       
            raise ValueError('Every side must be less than the sum of other sides!')
        if sum(self.angles) != 180:
            raise ValueError("The sum of the angles must be equal to 180 degrees!")
        max_angle_index = self.angles.index(max(self.angles))  #,       
        max_side_index = self.sides.index(max(self.sides))
        if max_angle_index != max_side_index:
            raise ValueError('The biggest angle must be opposite the biggest side')

    def get_info(self):
        return {
            'Name': self.name,
            'Angles': self.angles,
            'Sides': self.sides,
            'Perimetr': self.get_perimetr(),
            'Square': self.get_sq()
        }
    def draw(self):
        a, b, c = self.sides
        A, B, C = self.angles
        x1, y1 = 0, 0  #  
        x2, y2 = a, 0  #   (  X)
        x3 = b * math.cos(math.radians(C))
        y3 = b * math.sin(math.radians(C))
        points = [(x1, y1), (x2, y2), (x3, y3), (x1, y1)]  #  
        xs, ys = zip(*points)

        plt.plot(xs, ys, marker='o')
        plt.fill(xs, ys, alpha=0.3)
        plt.title("Triangle")
        plt.axis('equal')
        plt.grid()
        plt.show()

class Quadrangle(Shapes):
    def __init__(self, angles=None, sides=None):
        if angles is None and sides is None:
            raise ValueError("Quadrangle must have sides or angles.")

        if sides is not None:
            if len(sides) != 4:
                raise ValueError("Quadrangle must have 4 sides")
            self.sides = sides
        else:
            self.sides = [0, 0, 0, 0]  #    ,   

        if angles is not None:
            if len(angles) != 4:
                raise ValueError("Quadrangle must have 4 angles.")
            self.angles = angles
            self.fill_missing_angle()  #   
            self.validate_quadrangle()
        else:
            self.angles = [0, 0, 0, 0]  #    ,   

        self.n_angles = 4
        super().__init__(self.n_angles, self.angles, self.sides)

    def fill_missing_angle(self):
        missing_angle_index = [i for i, angle in enumerate(self.angles) if angle < 0]    #   ,    -1
        if len(missing_angle_index) == 1:
            current_sum = sum(angle for angle in self.angles if angle >= 0)
            missing_index = missing_angle_index[0]
            self.angles[missing_index] = 360 - current_sum

    def get_diagonals(self):
        di1 = math.sqrt(self.sides[0] ** 2 + self.sides[2] ** 2 - 2 * self.sides[0] * self.sides[2] * math.cos(
            math.radians(self.angles[1])))  #   
        di2 = math.sqrt(self.sides[1] ** 2 + self.sides[3] ** 2 - 2 * self.sides[1] * self.sides[3] * math.cos(
            math.radians(self.angles[3])))  #   
        angD = math.sin(math.radians(180.0 - self.angles[2]) ) #    (,   )
        return di1, di2, angD

    def get_sq(self):
        if len(self.sides) != 4:  #    
            raise ValueError("For calculating square we must know 4 sides.")
        d1 = math.sqrt(self.sides[0] ** 2 + self.sides[2] ** 2 - 2 * self.sides[0] * self.sides[2] * math.cos(math.radians(self.angles[1])))   #   
        d2 = math.sqrt(self.sides[1] ** 2 + self.sides[3] ** 2 - 2 * self.sides[1] * self.sides[3] * math.cos(math.radians(self.angles[3])))   #   
        angle_between_diagonals = math.radians(180.0 - self.angles[2])      #    (,   )
        area = 0.5 * d1 * d2 * math.sin(angle_between_diagonals)

        return area

    def set_angles(self, angles):
        if len(angles) != 4 or any(angle < 0 or angle >= 360 for angle in angles):
            raise ValueError("Quadrangle must have 4 positive angles less than 360.")
        self.angles = angles
        self.fill_missing_angle()
        self.validate_quadrangle()

    def set_sides(self, sides):
        if len(sides) != 4 or any(side <= 0 for side in sides):
            raise ValueError("Quadrangle must have 4 positive sides.")
        self.sides = sides

    def validate_quadrangle(self):
        if sum(self.angles) != 360:
            raise ValueError("The sum of angles must be equal 360 degrees.")

    def get_info(self):
        return {
            'Name': "Quadrangle",
            'Angles': self.angles,
            'Sides': self.sides,
            'Perimetr': self.get_perimetr(),
            'Area': self.get_sq(),
            'Diagonal 1, diagonal2, sin of angle between them' : self.get_diagonals()
        }
    def draw(self):
        if len(self.sides) != 4 or len(self.angles) != 4:
            print("There is not enough data to draw the shape.")
            return

        points = [(0, 0)]
        angle_offset = 0

        for i in range(4):
            angle_offset += self.angles[i]
            x = points[i][0] + self.sides[i] * math.cos(math.radians(angle_offset))
            y = points[i][1] + self.sides[i] * math.sin(math.radians(angle_offset))
            points.append((x, y))

        points.append(points[0])  #  
        xs, ys = zip(*points)

        plt.plot(xs, ys, marker='o')
        plt.fill(xs, ys, alpha=0.3)
        plt.title(self.name)
        plt.axis('equal')
        #plt.grid()
        plt.show()

class Nangle(Shapes):
    def __init__(self, angles, sides):
        super().__init__(len(angles), angles, sides)
        self.validate_nangles()

    def validate_nangles(self):
        expected_sum = (self.n_angles - 2) * 180
        actual_sum = sum(self.angles)

        if actual_sum != expected_sum:
            raise ValueError(f"The sum must be {expected_sum}. You have: {actual_sum}.")

    def get_sq(self):
        #  n-  
        raise NotImplementedError("Area calculation for n-angle is not implemented.")

    def get_info(self):
        return {
            'Name': f'N-angle with {self.n_angles} sides',
            'Angles': self.angles,
            'Sides': self.sides,
            'Perimetr': self.get_perimetr(),
        }

# 

circle = Circle(radius = 5)
print(circle.get_info())
circle.draw()

triangle = Triangle( sides = [3*math.sqrt(2), 6,  3*math.sqrt(2)])
triangle.set_sides([5, 5, 5])
print(triangle.get_info())
triangle.draw()

quadrangle = Quadrangle(angles=[90, 90, 90, 90], sides=[10, 10, 10, 10])
print(quadrangle.get_info())

quadrilateral = Quadrilateral(angles=[90, 90, 90, 90], sides=[10, 10, 10, 10])
print(quadrilateral.get_info())
quadrilateral.draw()

nangle = Nangle(angles = [108, 108, 108, 108, 108], sides = [5, 5, 5, 5, 5])
print(nangle.get_info())
```
